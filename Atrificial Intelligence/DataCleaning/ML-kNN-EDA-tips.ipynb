```python
# plot of 1 variable
sns.violinplot(x=tips['tip'])
```

```python
# plot of 1 variable
sns.violinplot(x=tips['total_bill'])
```

```python
# plot of 2 variables , size vs total_bill
plt.figure(figsize=(12,8))
sns.violinplot(x='size', y='total_bill', data=tips)
```

```python
sns.violinplot(x='day', y='total_bill', data=tips)
```

```python
# with hue
plt.figure(figsize=(12,8))
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex')

# more females pay tips
```

```python
# with hue
plt.figure(figsize=(12,8))
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker')

# smokers pay less total bills
```

```python
# with hue
plt.figure(figsize=(12,8))
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', split=True)

# smokers pay less total bills
```

```python
# with hue
plt.figure(figsize=(12,8))
sns.violinplot(x='day', y='total_bill', data=tips, hue='smoker', \
               split=True, inner='quartile')

# 3 lines
# top line - Q3 (above it are 25% of the values)
# middle line - Q2 Median (above/below it are 50% of the values)
# bottom line - Q1 (below it 25% of the values)
# 
```

```python
# with hue
plt.figure(figsize=(12,8))
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', \
               split=True, inner='quartile', scale='count')

# number of males and females
```

```python
# with hue
plt.figure(figsize=(12,8))
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', \
               inner='quartile', scale='count')

# number of males and females
```

```python
# with hue
plt.figure(figsize=(12,8))
sns.violinplot(x='day', y='total_bill', data=tips, hue='sex', \
               inner='stick', scale='count')

# number of males and females
```
